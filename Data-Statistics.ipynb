```python
def vector_addition(*args): # 패킹
    return [sum(t) for t in zip(*args)] # 언패킹
```

```python
u = [2, 2]
v = [1, 2]
z = [-3, 5]
print(vector_addition(u, v, z))
```

```python
vectors = [[2, 2,], [1, 2], [-3, 5]]
print(vector_addition(*vectors)) # 바로 언패킹해서 출력
```

```python
#Scala + Vector
u = [1, 2, 3]
v = [4, 3, 3]
alpha = 2
```

```python
# zip 내장함수
result = [alpha * sum(t) for t in zip(u, v)]
print(result)
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(-4.5, 5, 0.5)
y1 = 2*x**2     # 2x+2
y2 = 5*x + 30   # 5x+30
y3 = 4*x**2 + 10 #4x^2+10

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.show()
```

```python
plt.plot(x, y1)
plt.figure()
plt.plot(x, y2)
plt.show()
```

```python
x = np.arange(0, 10, 0.1)
y1 = 0.3*(x-5)**2 + 1
y2 = -1.5*x + 3
y3 = np.sin(x)**2
y4 = 10*np.exp(-x) + 1
```

```python
plt.subplot(2,2,1) #subplot 
plt.plot(x, y1)
plt.subplot(2,2,2)
plt.plot(x, y2)
plt.subplot(2,2,3)
plt.plot(x, y3)
plt.subplot(2,2,4)
plt.plot(x, y4)
plt.show()
```

```python
plt.plot(x, y1, 'm', x, y2, 'y', x, y3, 'k', x, y4, 'c')
plt.show() #''사이에 값 넣어줘서 그래프 색 지정
```

```python
plt.plot(x, y1, '-', x, y2, '--', x, y3, ':', x, y4, '-.')
plt.show() # ''사이에 값 넣어줘서 그래프 모양 지정
```

```python
plt.plot(x, y1, 'o', x, y2, '^', x, y3, 's', x, y4, 'd')
plt.show() #마커
```

```python
x = np.arange(0, 5, 1)
y1 = x
y2 = x + 1
y3 = x + 2
y4 = x + 3
```

```python
# ''사이의 값 순서는 항상 컬러, 스타일, 마커 순서로
plt.plot(x, y1, '>--r', x, y2, 's-g', x, y3, 'd:b', x, y4, '-.Xc')
plt.show()
```
